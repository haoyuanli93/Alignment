```python
import h5py as h5
import numpy as np
import matplotlib.pyplot as plt
```

```python
iou_holder = []
volume_holder = []
direction_holder = []

with h5.File('../output/gradient_output_cat_2_2018_08_06_14_27_28.h5','r') as h5file:
    print(list(h5file.keys()))
    
    # Load the two objects
    fixed = h5file['fixed_target'].value
    fixed_target = np.load(fixed)
    movable = h5file['movable_target'].value
    movable_target = np.load(movable)
    
    # Load the searching results
    for l in range(5):
        iou_holder.append(h5file['step_{}'.format(l) +'/iou' ].value)
        volume_holder.append(h5file['step_{}'.format(l)+'/volume'].value)
        direction_holder.append(h5file['step_{}'.format(l)+'/axis'].value)
        #direction_holder.append(h5file['step_{}'.format(l)+'/shift'].value)
```

```python
print(np.sum(fixed_target), np.sum(movable_target))
print(np.array(iou_holder)/iou_holder[0])
```

```python
fig, axes = plt.subplots(ncols=2)
fig.set_figwidth(10)
axes[0].imshow(fixed_target[13])
axes[1].imshow(movable_target[13])

plt.show()
```

```python
fig, axes = plt.subplots(ncols=5)
fig.set_figwidth(15)
for l in range(5):
    axes[l].imshow(volume_holder[l][13])

plt.show()
```
