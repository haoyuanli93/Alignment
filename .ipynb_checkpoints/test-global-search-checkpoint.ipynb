```python
import numpy as np
import matplotlib.pyplot as plt
import util
import scipy
import scipy.ndimage
```

# Define parameters

```python
delta_degree = 0.15
translation_range = 4

directions, _, dist_min = util.sample_full_2d_sphere(delta_theta=delta_degree, delta_phi=delta_degree)
degrees = util.sample_full_circle(delta_psi=delta_degree)
shift_list = np.arange(start = -4, stop=4)
```

# Load IoU list

```python
IoU_list = np.load('/reg/d/psdm/amo/amo86615/res/haoyuan/alignment/output/IoU_all_1.npy')
print(IoU_list.shape[0]/1e6)
print(IoU_list.max(), IoU_list.min())
```

```python
index = np.argmax(IoU_list)
print(index)
```

# Get the transformation

```python
axis, angle, shift = util.recover_the_transform(index,directions, degrees, shift_list,shift_list,shift_list)
```

# Reproduce the transformation

```python
# All node load the two arrays to compare
fixed_target = np.load('/reg/d/psdm/amo/amo86615/res/haoyuan/alignment/input/chop_complete.npy')
movable_target = np.load('/reg/d/psdm/amo/amo86615/res/haoyuan/alignment/input/chop_category_1.npy')

# Obtain the center of mass of the movable target
center = util.get_mass_center(movable_target)

# Calculate the affine map to use scipy.ndimage.affine_transform
rotation_matrix = util.angle_axis_to_mat(axis=axis, theta=angle)
offset = center - rotation_matrix.dot(center)

# Rotate the sample space
rotated = scipy.ndimage.affine_transform(input=movable_target,
                                           matrix=rotation_matrix,
                                           offset=offset,
                                           order=1,
                                           mode='constant', cval=0.0, prefilter=True)

# shift
shifted = scipy.ndimage.interpolation.shift(input=rotated,
                                              shift=shift,
                                              order=1,
                                              mode='constant',
                                              cval=0.0,
                                              prefilter=True)
```

```python
fig, axes  =plt.subplots(ncols=3, nrows=1)

fig.set_figwidth(12)

axes[0].imshow(np.sum(shifted, axis=0), vmax=0.01, cmap='jet')
axes[1].imshow(np.sum(shifted, axis=1), vmax=0.01, cmap='jet')
axes[2].imshow(np.sum(shifted, axis=2), vmax=0.01, cmap='jet')

plt.show()
```

```python
fig, axes  =plt.subplots(ncols=3, nrows=1)

fig.set_figwidth(12)

axes[0].imshow(np.sum(fixed_target, axis=0), vmax=0.01, cmap='jet')
axes[1].imshow(np.sum(fixed_target, axis=1), vmax=0.01, cmap='jet')
axes[2].imshow(np.sum(fixed_target, axis=2), vmax=0.01, cmap='jet')

plt.show()
```
