```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load two volumes

```python
# All node load the two arrays to compare
fixed_target = np.load('/reg/d/psdm/amo/amo86615/res/haoyuan/alignment/input/chop_complete.npy')
search_result = np.load('/reg/d/psdm/amo/amo86615/res/haoyuan/alignment/output/aligned_0_fine.npy')
```

```python
fig, axes  =plt.subplots(ncols=3, nrows=1)

fig.set_figwidth(12)

axes[0].imshow(np.sum(search_result, axis=0), vmax=0.01, cmap='jet')
axes[1].imshow(np.sum(search_result, axis=1), vmax=0.01, cmap='jet')
axes[2].imshow(np.sum(search_result, axis=2), vmax=0.01, cmap='jet')

plt.show()
```

```python
fig, axes  =plt.subplots(ncols=3, nrows=1)

fig.set_figwidth(12)

axes[0].imshow(np.sum(fixed_target, axis=0), vmax=0.01, cmap='jet')
axes[1].imshow(np.sum(fixed_target, axis=1), vmax=0.01, cmap='jet')
axes[2].imshow(np.sum(fixed_target, axis=2), vmax=0.01, cmap='jet')

plt.show()
```

```python
fig, axes  =plt.subplots(ncols=3, nrows=1)

fig.set_figwidth(12)

axes[0].imshow(search_result[13,:,:], vmax=0.01, cmap='jet')
axes[1].imshow(search_result[:,13,:], vmax=0.01, cmap='jet')
axes[2].imshow(search_result[:,:,13], vmax=0.01, cmap='jet')

plt.show()
```

```python
fig, axes  =plt.subplots(ncols=3, nrows=1)

fig.set_figwidth(12)

axes[0].imshow(fixed_target[13,:,:], vmax=0.01, cmap='jet')
axes[1].imshow(fixed_target[:,13,:], vmax=0.01, cmap='jet')
axes[2].imshow(fixed_target[:,:,13], vmax=0.01, cmap='jet')

plt.show()
```
