```python
import numpy as np
import matplotlib.pyplot as plt
```

# Load two volumes

```python
# All node load the two arrays to compare
fixed_target = np.load('/reg/d/psdm/amo/amo86615/res/haoyuan/alignment/input/chop_complete.npy')
search_result = np.load('/reg/d/psdm/amo/amo86615/res/haoyuan/alignment/output/aligned_0_coarse.npy')
```
